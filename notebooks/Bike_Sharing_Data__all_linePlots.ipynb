```python
import pandas as pd
pd.plotting.register_matplotlib_converters()    #   to convert pandas' datetime64 - should not be necessary with modern versions
import matplotlib.pyplot as plt
# to render the plot inside the jupyter notebook, instead of inside a pop-up window
%matplotlib inline
import seaborn as sns

```

```python
path1 = "../data/day.csv"
path2 = "../data/hour.csv"

# df = pd.read_csv(path1, index_col=1, parse_dates=True)      # day
df = pd.read_csv(path2, index_col=1, parse_dates=True)      # hour
# numeric_cols = df.select_dtypes(include='number').columns.toList()
numeric_cols = df.select_dtypes(include='number').columns
# print(numeric_cols)
# for e in numeric_cols:
#     print (e)
fig, axes = plt.subplots(4, 4 , figsize=(15, 15))
i = 0; j = 0
for col in numeric_cols:
    sns.lineplot(df[col], ax=axes[i, j])
    i += 1
    if i == 4:
        j += 1
        i = 0
```

```python
sns.barplot(data=df, x="hr", y="cnt")
```

```python
sns.boxplot(data=df, x='weekday', y='casual')
# sns.boxplot(data=df, x='weekday', y='registered')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.scatterplot(data=df, x='temp', y='cnt', ax=axes[0])
sns.regplot(data=df, x='temp', y='cnt', ax=axes[1])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16,5))
# fig, axes = plt.subplots(1, 3)
sns.lineplot(data=df.temp, ax=axes[0])
sns.lineplot(data=df.atemp, ax=axes[1])
sns.lineplot(data=df.atemp - df.temp, ax=axes[2])

```
