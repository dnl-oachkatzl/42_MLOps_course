```python
import pandas as pd
import numpy as np
from IPython.core.pylabtools import figsize

pd.plotting.register_matplotlib_converters()    #   to convert pandas' datetime64 - should not be necessary with modern versions
import matplotlib.pyplot as plt
# to render the plot inside the jupyter notebook, instead of inside a pop-up window
%matplotlib inline
import seaborn as sns
```

```python
path1 = "../data/day.csv"
path2 = "../data/hour.csv"

df = pd.read_csv(path1, index_col=1, parse_dates=True)
```

```python
correlation_matrix = df.corr(numeric_only=True)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
print(mask)
```

some text as commentary

```python
plt.figure(figsize=(16,16))
sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt=".2f", cmap='coolwarm')
# sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
# sns.heatmap(correlation_matrix)
```
